```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    df_sentiment = pd.read_csv('sentiment.csv')
    df_trades = pd.read_csv('trades.csv')
    print("‚úÖ Files loaded successfully!")
except FileNotFoundError:
    print("‚ùå ERROR: Please upload 'sentiment.csv' and 'trades.csv'.")

df_sentiment.columns = df_sentiment.columns.str.strip()
df_trades.columns = df_trades.columns.str.strip()

print(f"Sentiment Columns: {list(df_sentiment.columns)}")
print(f"Trader Columns: {list(df_trades.columns)}")

if 'date' in df_sentiment.columns:
    df_sentiment['Date'] = pd.to_datetime(df_sentiment['date'])
elif 'timestamp' in df_sentiment.columns:
    df_sentiment['Date'] = pd.to_datetime(df_sentiment['timestamp'])
else:
    df_sentiment.rename(columns={df_sentiment.columns[0]: 'Date'}, inplace=True)
    df_sentiment['Date'] = pd.to_datetime(df_sentiment['Date'])

if 'Timestamp' in df_trades.columns:
    df_trades['datetime'] = pd.to_datetime(df_trades['Timestamp'], errors='coerce')
    if df_trades['datetime'].dt.year.mean() < 2000:
        df_trades['datetime'] = pd.to_datetime(df_trades['Timestamp'], unit='ms')
elif 'Timestamp IST' in df_trades.columns:
     df_trades['datetime'] = pd.to_datetime(df_trades['Timestamp IST'])
else:
    raise KeyError("Could not find 'Timestamp' column in trades file.")

df_trades['date_only'] = df_trades['datetime'].dt.floor('d')

if 'Closed PnL' in df_trades.columns:
    df_trades.rename(columns={'Closed PnL': 'closedPnL'}, inplace=True)

df_merged = pd.merge(df_trades, df_sentiment, left_on='date_only', right_on='Date', how='inner')

if 'classification' in df_merged.columns:
    sent_col = 'classification'
elif 'Classification' in df_merged.columns:
    sent_col = 'Classification'
else:
    sent_col = 'value_classification'

def clean_sentiment(val):
    val = str(val).lower()
    if 'fear' in val: return 'Fear'
    if 'greed' in val: return 'Greed'
    return 'Neutral'

df_merged['Sentiment_Group'] = df_merged[sent_col].apply(clean_sentiment)

print(f"‚úÖ Data Merged! Analyzing {len(df_merged)} trades.")

df_merged['is_profit'] = df_merged['closedPnL'] > 0
win_rate = df_merged.groupby('Sentiment_Group')['is_profit'].mean() * 100
avg_pnl = df_merged.groupby('Sentiment_Group')['closedPnL'].mean()

sns.set_style("whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.barplot(x=win_rate.index, y=win_rate.values, ax=axes[0], palette="RdYlGn")
axes[0].set_title('Win Rate (%) by Sentiment')
axes[0].set_ylabel('Win Rate (%)')

sns.barplot(x=avg_pnl.index, y=avg_pnl.values, ax=axes[1], palette="coolwarm")
axes[1].set_title('Average PnL ($) by Sentiment')
axes[1].set_ylabel('Avg Profit/Loss ($)')

plt.tight_layout()
plt.show()

print("\n--- üìä SUMMARY ---")
print("Win Rate:\n", win_rate)
print("\nAvg PnL:\n", avg_pnl)
```
